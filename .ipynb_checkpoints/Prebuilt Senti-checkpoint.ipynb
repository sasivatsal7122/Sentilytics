# VADER

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
nltk.download('vader_lexicon')
%matplotlib inline
import warnings
from nltk.sentiment.vader import SentimentIntensityAnalyzer

warnings.filterwarnings("ignore")

import os
```

```python
comments = pd.read_csv('3MZ6aTpFrE0_Afiltered.csv', error_bad_lines=False)
```

```python
comments
```

```python
comments['Comments'] = comments['Comments'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
comments['Comments'] = comments['Comments'].apply(lambda x:x.lower())

tokenized_tweet = comments['Comments'].apply(lambda x: x.split())

wnl = WordNetLemmatizer()

tokenized_tweet.apply(lambda x: [wnl.lemmatize(i) for i in x if i not in set(stopwords.words('english'))]) 
tokenized_tweet.head()

for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])
    
    
comments['Comments'] = tokenized_tweet

sia = SentimentIntensityAnalyzer()
comments['Sentiment Scores'] = comments['Comments'].apply(lambda x:sia.polarity_scores(x)['compound'])
comments['Sentiment'] = comments['Sentiment Scores'].apply(lambda s : 'Positive' if s > 0 else ('Neutral' if s == 0 else 'Negative'))
comments.Sentiment.value_counts()

percentages = comments['Sentiment'].value_counts(normalize=True) * 100

vader_neutral_df = comments[comments['Sentiment'] == 'Neutral']
vader_positive_df = comments[comments['Sentiment'] == 'Positive']
vader_negative_df = comments[comments['Sentiment'] == 'Negative']

print(percentages)
```

# Text Blob

```python
from textblob import TextBlob
import pandas as pd

df = pd.read_csv('3MZ6aTpFrE0_Afiltered.csv')

textBlob_df_positive = pd.DataFrame(columns=df.columns)
textBlob_df_negative = pd.DataFrame(columns=df.columns)
textBlob_df_neutral = pd.DataFrame(columns=df.columns)
for index, row in df.iterrows():
    comment = row['Comments']
    
    blob = TextBlob(comment)
    polarity = blob.sentiment.polarity
    
    if polarity > 0:
        textBlob_df_positive = textBlob_df_positive.append(row)
    elif polarity < 0:
        textBlob_df_negative = textBlob_df_negative.append(row)
    else:
        textBlob_df_neutral = textBlob_df_neutral.append(row)

# Optionally, you can save the three DataFrames to separate CSV files
# textBlob_df_positive.to_csv('positive_comments.csv', index=False)
# textBlob_df_negative.to_csv('negative_comments.csv', index=False)
# textBlob_df_neutral.to_csv('neutral_comments.csv', index=False)
```

# Xlnet

```python
import pandas as pd
from transformers import XLNetTokenizer, XLNetForSequenceClassification

data = pd.read_csv("3MZ6aTpFrE0_Afiltered.csv")

# Load the XLNet tokenizer and model
tokenizer = XLNetTokenizer.from_pretrained("xlnet-base-cased")
model = XLNetForSequenceClassification.from_pretrained("xlnet-base-cased", num_labels=3)

# Function to classify sentiment using XLNet
def get_sentiment(text):
    inputs = tokenizer.encode_plus(text, add_special_tokens=True, padding="max_length", truncation=True, max_length=128, return_tensors="pt")
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]
    outputs = model(input_ids, attention_mask=attention_mask)
    logits = outputs.logits
    predicted_labels = logits.argmax(dim=1)
    sentiment_labels = ["Negative", "Neutral", "Positive"]
    sentiment = sentiment_labels[predicted_labels.item()]
    return sentiment

# Apply sentiment analysis to the 'Comments' column
data["Sentiment"] = data["Comments"].apply(get_sentiment)

# Split data into separate dataframes based on sentiment
xlent_positive_df = data[data["Sentiment"] == "Positive"]
xlent_negative_df = data[data["Sentiment"] == "Negative"]
xlent_neutral_df = data[data["Sentiment"] == "Neutral"]
```

# AFINN Model

```python
import pandas as pd
from afinn import Afinn

data = pd.read_csv("3MZ6aTpFrE0_Afiltered.csv")

afinn = Afinn()

def get_sentiment(text):
    sentiment_score = afinn.score(text)
    if sentiment_score > 0:
        return "Positive"
    elif sentiment_score < 0:
        return "Negative"
    else:
        return "Neutral"

data["Sentiment"] = data["Comments"].apply(get_sentiment)

afinn_positive_df = data[data["Sentiment"] == "Positive"]
afinn_negative_df = data[data["Sentiment"] == "Negative"]
afinn_neutral_df = data[data["Sentiment"] == "Neutral"]
```
